## New_df

```python
# 최적화
import numpy as np

def extract_keywords(df, keywords):
    mask = np.column_stack([df['text'].str.contains(f"{kw}", na=False) for kw in keywords])
    df_ex = df.loc[mask.any(axis=1)]
    return df_ex

keywords = ['Newton', 'system', 'emblem', 'diffraction', 'reflection', 'refraction', 'interference', 'electric', 'magnet',
            'matrix', 'theor', 'fuction', 'equation', 'model', 'philosopher', 'ultraviolet', 'visible', 'infrared',
            'ray', 'effect', 'physic', 'chemist', 'geometry', 'galaxy', 'Nobel Prize', 'neutron', 'electron', 'quantum',
            'collapse', 'scale', 'momentum', 'symmetr', 'definition', 'principle', 'solution', 'matter', 'spin', 'shift', 'space', 'orbit',
            'cosmo', 'astro', 'universe', 'force', 'particle', 'Einstein', 'measure', 'science', 'material', 'energy',
            'light', 'mathmatic', 'dimension', 'level', 'metal', 'origin', 'wave', 'motion', 'relationship', 'sinusoidal',
            'cosine', 'tangent', 'fluid', 'cell', 'plant', 'disease', 'plane', 'phenomenon', 'temperature', 'operat', 'radio',
            'mass', 'experiment', 'process', '17th', '20th', 'molecule', 'carbo', 'method', 'probability', ]
```

```python
new_df = extract_keywords(df, keywords)
new_df.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=True)
new_df = new_df.dropna().reset_index(drop=True)
new_df.info()
```

```python
newy_df = extract_keywords(df, keywords)
newy_df.drop_duplicates(subset=None, keep='first', inplace=True, ignore_index=True)
newy_df = newy_df.dropna().reset_index(drop=True)
newy_df.info()
```
